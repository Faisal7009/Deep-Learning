# Updated Code on CNN with Fruits Image Datasets [MYcode.V2]

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
```

```python
from sklearn.model_selection import train_test_split
import os
```

```python
np.random.seed(1)
```

```python
train_images = []
train_labels = []

shape = (200, 200)
train_datasets_path = "datasets/train_zip/train/"
for filename in os.listdir("datasets/train_zip/train/"):
    if filename.split('.')[1] =='jpg':
        img = cv2.imread(os.path.join(train_datasets_path, filename))
        train_labels.append(filename.split('_')[0])
        img = cv2.resize(img,shape)
        train_images.append(img)

train_labels = pd.get_dummies(train_labels).values
train_images = np.array(train_images)
```

```python
train_images = train_images.astype('float32')
train_images = train_images/255
```

```python
x_train, x_test, y_train, y_test = train_test_split(train_images, train_labels, test_size=0.33, random_state=1)
```

```python
x_train
```

# Models

```python
model = Sequential()

# 1st / Input Layer
model.add(Conv2D(kernel_size = (3, 3), filters = 32, padding = 'same', activation = 'relu', input_shape = (200, 200, 3)))
model.add(MaxPooling2D((2, 2),strides = 2))

# 2nd Hidden Layer
model.add(Conv2D(filters = 32, kernel_size = (3, 3), padding = 'same', activation = 'relu'))
model.add(MaxPooling2D((2, 2),strides = 2))

model.add(Dropout(0.5))
# flat layer
model.add(Flatten())

model.add(Dense(128, activation = 'relu'))
model.add(Dense(4, activation = 'softmax'))
```

```python
model.compile(optimizer = 'adam',
              loss = 'categorical_crossentropy',
              metrics = ['accuracy'])
```

```python
model.summary()
```

```python
model_data  = model.fit(x_train,
                        y_train,
                        epochs=5,
                        batch_size = 50,
                        validation_data=(x_test, y_test))
```

```python
plt.plot(model_data.history['accuracy'])
plt.plot(model_data.history['val_accuracy'])
plt.title("Model Accuracy")
plt.xlabel('epochs')
plt.ylabel("accuracy")
plt.tight_layout()
plt.legend(['train', 'test'], loc = 'upper left')
plt.show()
```

```python
plt.plot(model_data.history['loss'])
plt.plot(model_data.history['val_loss'])
plt.title("Model Loss")
plt.xlabel('epochs')
plt.ylabel("loss")
plt.tight_layout()
plt.legend(['train', 'test'], loc = 'upper left')
plt.show()
```
